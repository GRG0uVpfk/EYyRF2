## Problem Description
The task is to perform clustering on the Iris dataset to identify natural groupings among the samples based on their features. This will help in understanding the inherent structure of the data without using the labels.


## Exploratory Data Analysis (EDA)


```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Display basic information about the dataset
print(iris_df.info())
print(iris_df.describe())

# Pairplot to visualize relationships between features
sns.pairplot(iris_df)
plt.show()
```

## Model Building and Training

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler()
iris_scaled = scaler.fit_transform(iris_df)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
iris_df['Cluster'] = kmeans.fit_predict(iris_scaled)

# Visualize the clustering
sns.scatterplot(x=iris_df['petal length (cm)'], y=iris_df['petal width (cm)'], hue=iris_df['Cluster'], palette='viridis')
plt.title('KMeans Clustering of Iris Dataset')
plt.show()
```

## Result and Discussion
The KMeans algorithm identified three clusters in the Iris dataset, which correspond to the three species of iris flowers. The clustering was performed without using the species labels, demonstrating the power of unsupervised learning to uncover patterns in the data.
